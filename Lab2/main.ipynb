```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# wczytaj obraz i konwertuj do skali szarości
img = Image.open("12.webp").convert("L")
A = np.array(img, dtype=float)
A = (A - np.mean(A)) / np.std(A)

R_rows = np.corrcoef(A)  # korelacja między wierszami
R_cols = np.corrcoef(A.T)  # korelacja między kolumnami
U, S, Vt = np.linalg.svd(A, full_matrices=False)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.title("Macierz korelacji wierszy")
plt.imshow(R_rows, cmap="hot", interpolation="nearest")
plt.colorbar()

plt.subplot(1, 2, 2)
plt.title("Macierz korelacji kolumn")
plt.imshow(R_cols, cmap="hot", interpolation="nearest")
plt.colorbar()

plt.show()
```
